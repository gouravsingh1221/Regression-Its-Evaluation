#Q1 What is Simple Linear Regression ?

<br>

**Ans.-**  Simple Linear Regression is a fundamental algorithm in Machine Learning that models the relationship between **two variables**:
- **Independent Variable (X):** The predictor or input feature.  
- **Dependent Variable (Y):** The target or output we want to predict.  

The relationship is represented by the equation of a straight line:

\[
Y = a + bX
\]

Where:  
- **a (Intercept):** The value of Y when X = 0  
- **b (Slope):** The change in Y for every 1-unit increase in X  

---

###  Example  
Suppose we want to predict the **selling price of a bike (Y)** based on its **age in years (X)**.  
If the slope (b) = -5000, it means the bike’s price decreases by ₹5000 for each additional year of age.  

---

###  Importance  
- Helps in **prediction** of continuous values.  
- Shows the **strength and direction** of relationships.  
- Forms the **basis for more advanced regression models** in Machine Learning.  

#Q2 What are the key assumptions of Simple Linear Regression?

<br>

**Ans.-** For Simple Linear Regression to give valid results, a few assumptions must be satisfied:

1. **Linearity**  
   - The relationship between the independent variable (X) and dependent variable (Y) is linear.  
   - Example: As bike age increases, price decreases in a roughly straight-line fashion.  

2. **Independence of Errors**  
   - The residuals (errors) should be independent of each other.  
   - This means one observation should not influence another.  

3. **Homoscedasticity (Constant Variance of Errors)**  
   - The variance of residuals should remain constant across all values of X.  
   - In simple words, the spread of errors should be roughly equal for small and large values of X.  

4. **Normality of Errors**  
   - The residuals should follow a normal distribution.  
   - This helps in making reliable predictions and valid hypothesis testing.  

5. **No Significant Outliers**  
   - Extreme values can distort the regression line.  
   - Outliers should be detected and handled before fitting the model.  

---

###  Why These Assumptions Matter  
If these assumptions are violated, the regression model may give **biased, unreliable, or misleading results**. Ensuring these conditions are met improves the accuracy and interpretability of predictions.  

#Q3  What is heteroscedasticity, and why is it important to address in regression models?

<br>

**Ans.-** **Heteroscedasticity** occurs when the variance of the residuals (errors) in a regression model is **not constant** across all levels of the independent variable.  

- In a well-behaved regression model, the residuals should have **equal spread** (constant variance), a condition called **homoscedasticity**.  
- When the spread increases or decreases with the value of X, it is called **heteroscedasticity**.  

---

###  Why It Matters
1. **Biased Standard Errors**  
   - Heteroscedasticity makes the estimated standard errors unreliable.  
   - This can lead to incorrect confidence intervals and p-values.  

2. **Unreliable Hypothesis Testing**  
   - T-tests and F-tests may become invalid, leading to wrong conclusions about variable significance.  

3. **Reduced Model Efficiency**  
   - The regression line may still be unbiased, but predictions are less precise.  

---

###  Example (Real-World)  
In bike price prediction, if cheaper bikes show small variation in prices while luxury bikes show very high variation, the model suffers from heteroscedasticity.  

---

###  Importance  
Addressing heteroscedasticity ensures that regression models are **reliable, accurate, and statistically valid**. Techniques like **log transformation**, **weighted regression**, or using **robust standard errors** are commonly applied to fix this issue.  

#Q4 What is Multiple Linear Regression?

<br>

**Ans.-** **Multiple Linear Regression (MLR)** is an extension of Simple Linear Regression that models the relationship between a **dependent variable (Y)** and **two or more independent variables (X₁, X₂, X₃, ... Xₙ)**.  

The general equation is:

\[
Y = a + b_1X_1 + b_2X_2 + b_3X_3 + \dots + b_nX_n
\]

Where:  
- **Y** → Dependent variable (target we want to predict)  
- **X₁, X₂, …, Xₙ** → Independent variables (predictors/features)  
- **a** → Intercept  
- **b₁, b₂, …, bₙ** → Coefficients showing how much Y changes with each predictor  

---

###  Example  
Suppose we want to predict the **selling price of a bike (Y)** using:  
- **Bike age (X₁)**  
- **Kilometers driven (X₂)**  
- **Engine power (X₃)**  

The regression equation could look like:  

\[
Price = 150000 - 5000 \times (Age) - 2 \times (Km\_Driven) + 2000 \times (Power)
\]

---

###  Importance  
- Handles **multiple factors at once**, giving a more realistic model.  
- Helps in **prediction** when outcomes depend on many variables.  
- Useful for **understanding the contribution** of each feature towards the target variable.  

#Q5 What is polynomial regression, and how does it differ from linear regression?

<br>

**Ans.-** **Polynomial Regression** is a type of regression analysis where the relationship between the independent variable (X) and the dependent variable (Y) is modeled as an **nth-degree polynomial**.  

The general equation is:

\[
Y = a_0 + a_1X + a_2X^2 + a_3X^3 + \dots + a_nX^n
\]

Where:  
- **Y** → Dependent variable (target)  
- **X** → Independent variable (predictor)  
- **a₀, a₁, …, aₙ** → Coefficients  
- **n** → Degree of the polynomial  

---

###  Difference from Linear Regression  
- **Linear Regression** fits a straight line to the data:  
  \[
  Y = a + bX
  \]  
  It assumes a **linear relationship** between X and Y.  

- **Polynomial Regression** allows for **curved relationships** by including higher powers of X (like \(X^2, X^3\), etc.).  
  This makes it suitable for datasets where the trend is not straight but curved.  

---

###  Example  
Suppose we are predicting a bike’s **resale price (Y)** based on its **age (X)**:  
- A **linear model** might suggest price decreases at a constant rate each year.  
- A **polynomial model** might capture that price drops quickly in the first few years and then levels off later.  

---

###  Key Point  
Polynomial Regression is more flexible than Linear Regression, but if the degree of the polynomial is too high, it may lead to **overfitting**.  

#Q6 Implement a Python program to fit a Simple Linear Regression model to the following sample data:
● X = [1, 2, 3, 4, 5]

● Y = [2.1, 4.3, 6.1, 7.9, 10.2]

## Plot the regression line over the data points.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)   # Independent variable (reshaped for sklearn)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])       # Dependent variable

# Step 1: Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Step 2: Predict values
Y_pred = model.predict(X)

# Step 3: Plot original data points
plt.figure(figsize=(7,5))
plt.scatter(X, Y, color='blue', label="Actual Data", s=60, edgecolor="black")

# Step 4: Plot regression line
plt.plot(X, Y_pred, color='red', linewidth=2, label="Regression Line")

# Step 5: Add labels and title
plt.title("Simple Linear Regression", fontsize=14, weight='bold')
plt.xlabel("X (Independent Variable)", fontsize=12)
plt.ylabel("Y (Dependent Variable)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# Step 6: Print model parameters
print("Intercept (a):", model.intercept_)
print("Slope (b):", model.coef_[0])
```

#Q7 Fit a Multiple Linear Regression model on this sample data:

● Area = [1200, 1500, 1800, 2000]

● Rooms = [2, 3, 3, 4]

● Price = [250000, 300000, 320000, 370000]

##Check for multicollinearity using VIF and report the results.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# sample data
area = [1200, 1500, 1800, 2000]
rooms = [2, 3, 3, 4]
price = [250000, 300000, 320000, 370000]

# create dataframe
data = pd.DataFrame({
    'Area': area,
    'Rooms': rooms,
    'Price': price
})

# independent and dependent variables
X = data[['Area', 'Rooms']]
y = data['Price']

# fit multiple linear regression model
model = LinearRegression()
model.fit(X, y)

# print coefficients
print("Intercept (a):", model.intercept_)
print("Coefficients (b1 for Area, b2 for Rooms):", model.coef_)

# check for multicollinearity using VIF
# we need to add a constant column for VIF calculation
X_vif = pd.DataFrame(X, columns=['Area', 'Rooms'])
vif_data = pd.DataFrame()
vif_data["Feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

print("\nVariance Inflation Factor (VIF) Results:")
print(vif_data)
```

#Q8 Implement polynomial regression on the following data:

● X = [1, 2, 3, 4, 5]


● Y = [2.2, 4.8, 7.5, 11.2, 14.7]

##Fit a 2nd-degree polynomial and plot the resulting curve.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# create polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# fit the polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# make predictions
Y_pred = model.predict(X_poly)

# plot actual data points
plt.figure(figsize=(7,5))
plt.scatter(X, Y, color='blue', label="Actual Data", s=60, edgecolor="black")

# plot polynomial regression curve
plt.plot(X, Y_pred, color='red', linewidth=2, label="Polynomial Curve (degree 2)")

# add labels, title and legend
plt.title("Polynomial Regression (2nd Degree)")
plt.xlabel("X (Independent Variable)")
plt.ylabel("Y (Dependent Variable)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# print coefficients
print("Intercept (a):", model.intercept_)
print("Coefficients (b1, b2):", model.coef_)
```

#Q9 Create a residuals plot for a regression model trained on this data:

● X = [10, 20, 30, 40, 50]

● Y = [15, 35, 40, 50, 65]

##Assess heteroscedasticity by examining the spread of residuals.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# fit simple linear regression model
model = LinearRegression()
model.fit(X, Y)

# make predictions
Y_pred = model.predict(X)

# calculate residuals
residuals = Y - Y_pred

# plot residuals
plt.figure(figsize=(7,5))
plt.scatter(X, residuals, color='blue', s=60, edgecolor="black")
plt.axhline(y=0, color='red', linestyle="--", linewidth=2)

# add labels and title
plt.title("Residuals Plot")
plt.xlabel("X (Independent Variable)")
plt.ylabel("Residuals (Y - Y_pred)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# print residuals for reference
print("Residuals:", residuals)
```
